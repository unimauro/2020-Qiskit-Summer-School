```
import urllib3
http = urllib3.PoolManager()
for j in range(1,11):
  url='https://b-ok.lat/s/quantum%20computing?order=year&page='+str(j)
  #print(url)
  res = http.request('GET', url)
  for i in range(1,51):
    print(
        res.data.decode('utf-8').split('resItemBox resItemBoxBooks exactMatch')[i].split('bookProperty property_year')[1].split('property_value')[1].split('>')[1].split('<')[0]+
        '-::-'+
        res.data.decode('utf-8').split('resItemBox resItemBoxBooks exactMatch')[i].split('text-decoration')[1].split('</a')[0].split('>')[1]+'-::-')
```
